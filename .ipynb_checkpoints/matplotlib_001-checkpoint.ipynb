次の例では
- 横軸に０以上１以下の乱数
- 縦軸に横軸*2+正規乱数(平均0標準偏差1の正規分布に従う乱数)

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
x=np.random.rand(100)
print(x)
y=2*x+np.random.rand(100)
plt.scatter(x,y)
plt.show()
```

```python
plt.rcParams['font.size']=24
plt.title("tttiiiiet")
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("ekkisu")
plt.ylabel("yai")
plt.show()
```

```python
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
plt.scatter(x,y,c="r")
plt.grid()
plt.show()
```

```python
plt.plot(np.random.rand(100),c='r',marker='o', linestyle='-', label='1')
plt.show()
```

```python
plt.plot(np.random.rand(100),c='r',marker='o', linestyle='-', label='fff')
plt.plot(np.random.rand(100),c='b',marker='v', linestyle='--', label='ggg')
plt.xlim(-1, 50)
plt.ylim(-1, 3.5)
plt.legend(loc='upper right', prop={'size':8})
plt.show()
```

```python
N=100

x=np.random.rand(N)
y=np.random.rand(N)
c=np.random.rand(N)

fig=plt.figure(figsize=(15,8))
fig=plt.figure()

plt.subplot(2,2,1)
plt.scatter(x,y)
plt.title("aaa")
# plt.colorbar()

plt.subplot(2,2,2)
plt.scatter(x,y,c=c)
plt.title("bbb")
plt.colorbar()

plt.subplot(2,2,3)
plt.scatter(x,y,c=c,cmap=plt.cm.Greens)
plt.title("ccc")
plt.colorbar()

plt.subplot(2,2,4)
plt.scatter(x,y,c=c,cmap=plt.cm.jet)
plt.title("ddd")
plt.colorbar()

plt.subplots_adjust(hspace=0.35)

plt.show()
```

```python
fig=plt.figure()
ax=fig.gca()
ax.plot(np.random.rand(100))
fig.show()
```

```python
np.random.seed(123)
N=100
x=np.random.rand(N)
y=np.random.rand(N)
plt.figure(figsize=(10,8))
plt.subplot(221)
plt.scatter(x,y,marker='o')
plt.subplot(222)
plt.scatter(x,y,marker='x')
plt.subplot(223)
plt.scatter(x,y,marker='s')
plt.subplot(223)
plt.scatter(x,y,marker='^')

plt.show()
```

```python
# %matplotlib nbagg
import matplotlib.animation as animation

N=100
np.random.seed(123)
x=np.random.randn(N)

absmax_x=np.ceil(np.max(np.abs(x)))
xmin, xmax = -absmax_x, absmax_x

def animate(nframe):
    plt.cla()
    plt.hist(x[:nframe],bins=20)
    plt.xlim(xmin,xmax)
    plt.ylim(0,100)
    plt.title('N=%s'%nframe)

fig=plt.figure()
anim=animation.FuncAnimation(fig,animate,np.arange(1,N+1))
plt.show()
```
